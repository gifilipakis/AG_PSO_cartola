```python
import pandas as pd
import random
```

```python
jogadores = pd.read_csv("caRtola/data/desafio_valorizacao/valorizacao_cartola_2018.csv")
jogadores = jogadores[['id','rodada','posicao','media_pontos','preco','status']].sample(n=2000, random_state=1)
jogadores.head()
```

```python
class Time:
    def __init__(self, formacao, time):
        self.formacao = formacao
        self.time = time
        self.preco = 0
        self.fit = 0
        self.pbest = self.time
        self.pbest_fit = self.fit
        self.velocidade = [0]*12
        self.w = 1.1
        
    def calc_fit(self,jogadores_filt):
        pontos = 0
        for v in self.time:
            pontos += jogadores_filt.iloc[v]['media_pontos']
        self.fit = pontos
        return self.fit

    def calc_custo(self):
        preco = 0
        for v in self.time:
            preco += jogadores_filt.iloc[v]['preco']
        self.preco = preco
        return self.preco
    
    def calc_velocidade(self,gbest):
        c1 = c2 = 2
        self.w -= 0.1
        tx_aprendizagem_1 = random.random()
        tx_aprendizagem_2 = random.random()
        for i in range(len(self.velocidade)):
            self.velocidade[i] = int(self.w*self.velocidade[i] + c1*tx_aprendizagem_1*(self.pbest[i]-self.time[i])\
            +c2*tx_aprendizagem_2*(gbest[i]-self.time[i]))
    
    def update_pos(self,vmax,gbest):
        for i in range(len(self.time)):
            if abs((self.time[i]+self.velocidade[i])-self.pbest[i]) <= abs((self.time[i]+self.velocidade[i])-gbest[i]):
                self.time[i] = self.pbest[i]
            else:
                self.time[i] = gbest[i]
                
#             if self.time[i] + self.velocidade[i] < 0:
#                 self.time[i] = 0
#             elif self.time[i] + self.velocidade[i] < vmax:
#                 self.time[i] += self.velocidade[i]
#             else:
#                 self.time[i] = vmax-1
    
def cria_particulas(formacao,jogadores_filt):
    particulas = []
    for i in range(5):
        time = []
        if formacao[0] == '4':
            zag = jogadores_filt.loc[jogadores_filt['posicao']=='zag'].sample(n = 2, replace = False)
            lat = jogadores_filt.loc[jogadores_filt['posicao']=='lat'].sample(n = 2, replace = False)
            time += zag['order'].tolist()+lat['order'].tolist()
        elif formacao[0] == '5':
            zag = jogadores_filt.loc[jogadores_filt['posicao']=='zag'].sample(n = 3, replace = False)
            lat = jogadores_filt.loc[jogadores_filt['posicao']=='lat'].sample(n = 2, replace = False)
            time += zag['order'].tolist()+lat['order'].tolist()
        else:
            defesa = jogadores_filt.loc[(jogadores_filt['posicao']=='zag')|(jogadores_filt['posicao']=='lat')].sample(n = 3, replace = False)
            time += defesa['order'].tolist()
        mei = jogadores_filt.loc[jogadores_filt['posicao']=='mei'].sample(n = int(formacao[1]), replace = False)
        time += mei['order'].tolist()
        
        ata = jogadores_filt.loc[jogadores_filt['posicao']=='ata'].sample(n = int(formacao[2]), replace = False)
        time += ata['order'].tolist()
        
        tec = jogadores_filt.loc[jogadores_filt['posicao']=='tec'].sample(n = 1, replace = False)
        time += tec['order'].tolist()
        
        gol = jogadores_filt.loc[jogadores_filt['posicao']=='gol'].sample(n = 1, replace = False)
        time += gol['order'].tolist()
        
        time_formado = Time(formacao,time)
        particulas.append(time_formado)
    return particulas

def retorna_time(formacao,num_rodada,num_cartoletas,max_iteracoes):
    jogadores_filt = jogadores.loc[(jogadores['rodada']==num_rodada) & ((jogadores['status'] != 'Contundido') | (jogadores['status'] != 'Suspenso'))]
    vmax = jogadores_filt['id'].count()
    jogadores_filt.insert(0, 'order', list(range(0,vmax)))
    gbest = None
    gbest_fit = 0
    particulas = cria_particulas(formacao,jogadores_filt)
    it = 0
    while it < max_iteracoes:
        for p in particulas:
            fit = p.calc_fit(jogadores_filt)
            if fit > gbest_fit:
                gbest_fit = fit
                gbest = p.time
                p.pbest = p.time
                p.pbest_fit = fit 
            elif fit > p.pbest_fit:
                p.pbest = p.time
                p.pbest_fit = fit 
        for p in particulas:
            p.calc_velocidade(gbest)
            p.update_pos(vmax,gbest)
        it += 1
    final_frame = jogadores_filt.loc[jogadores_filt['order'].isin(gbest)]
    return [gbest_fit,final_frame]

def PSO(num_cartoletas,num_rodadas,max_iteracoes):
    results = []
    for num_rodada in num_rodadas:
        max_team = [0,0]
        for1 = retorna_time('433',num_rodada,num_cartoletas,max_iteracoes)
        for2 = retorna_time('442',num_rodada,num_cartoletas,max_iteracoes)
        for3 = retorna_time('451',num_rodada,num_cartoletas,max_iteracoes)
        for4 = retorna_time('352',num_rodada,num_cartoletas,max_iteracoes)
        for5 = retorna_time('343',num_rodada,num_cartoletas,max_iteracoes)
        for6 = retorna_time('532',num_rodada,num_cartoletas,max_iteracoes)
        for7 = retorna_time('541',num_rodada,num_cartoletas,max_iteracoes)
        
        teams = [for1,for2,for3,for4,for5,for6,for7]
        for t in teams:
            if t[0] > max_team[0]:
                max_team = t
        results.append(max_team)
    return results

pso = PSO(100,[11,30,38],200)
for p in pso:
    print('Pontuação Final:',p[0])
    print(p[1])
    print('---------------------------------------------------------------------------')
```
